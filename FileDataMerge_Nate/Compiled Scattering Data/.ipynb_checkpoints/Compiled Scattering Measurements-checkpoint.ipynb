```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
myPath = input("Enter Directory: ")  #To get rid of text at beginning of ran cell, replace " " with Directory Route
lambdas = []
counts = []

def plotData(lambdas, counts, item):
    plt.plot(lambdas, counts)
    plt.title(item)
    plt.xlabel('Wavelength [nm]')
    plt.ylabel('Counts')
    plt.show()
    
def parseData(line):
    data = line.split('\t')
    lambda0 = float(data[0])
    count = float(data[1].replace('\n',''))
    lambdas.append(lambda0)
    counts.append(count)
            
def read_file(FilePath):
    file = open(FilePath, 'r')
    line = file.readline()
    while line:
        if not line.startswith('#'):
            parseData(line)
        line = file.readline()

def myDirectoryPrinter(thePath):
    for item in os.listdir(thePath):
        if not os.path.isdir(os.path.join(thePath, item)):
            if item.endswith(".egg"):
                read_file(os.path.join(thePath, item))
                global lambdas
                global counts 
                plotData(lambdas,counts,item)
                lambdas = []
                counts = []
        else: 
            myDirectoryPrinter(os.path.join(thePath, item))

                    
myDirectoryPrinter(myPath)
```
